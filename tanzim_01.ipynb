```python
import json
import os
import numpy as np
import cv2
from PIL import Image as PILImage
from matplotlib import pyplot as plt
from datasets import Dataset, Features, Image, Value
```

```python
import json

def read_json_file(file_path):
  with open(file_path, 'r') as file:
    data = json.load(file)
  return data

# Example usage (assuming 'example.json' is in the current directory)
file_path = '/content/rec_dataset_v1/train.json'  # Replace with your JSON file path
data = read_json_file(file_path)

# Now you can work with the data loaded from the JSON file
# For example, print the data
len(data)
```

```python
import random

# Randomly select 3600 samples
random_samples = random.sample(data, 4500)

# Verify the size of the selected data
print(f"Number of selected samples: {len(random_samples)}")
```

```python
dataset = Dataset.from_list(data[:2900])

# Verify dataset
print(dataset)
```

```python
def load_image(example):
    try:
        example["image"] = PILImage.open(example["image_path"]).convert("RGB")
        return example
    except FileNotFoundError:
        print(f"Error: Image file not found at '{example['image_path']}'")
        return None  # or handle the error differently



# Apply the image loading function
dataset = dataset.map(load_image, remove_columns=["image_path"])

# Filter out any examples where image loading failed
dataset = dataset.filter(lambda example: example is not None)

# Verify dataset
print(dataset)
```

```python
from transformers import TrOCRProcessor
import torch
```

```python
from transformers import TrOCRProcessor
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
# Load TrOCR processor
processor = TrOCRProcessor.from_pretrained("/content/drive/MyDrive/OCR-Custom-Dataset/trocr-rec-model-pre")

# Preprocessing function
def preprocess(example):
    # Resize and convert the image to pixel values
    example["pixel_values"] = processor(images=example["image"], return_tensors="pt").pixel_values[0].to(device)
    # Tokenize the text labels
    example["labels"] = processor.tokenizer(example["text"],
                                            padding="max_length",
                                            truncation=True,
                                            return_tensors="pt").input_ids[0].to(device)
    return example

# Apply preprocessing to the dataset
processed_dataset = dataset.map(preprocess, remove_columns=["image", "text"])
```

```python
train_test_split = processed_dataset.train_test_split(test_size=0.01)
train_dataset = train_test_split["train"]
val_dataset = train_test_split["test"]

print(f"Train set: {len(train_dataset)}, Validation set: {len(val_dataset)}")
```

```python
from transformers import VisionEncoderDecoderModel

# Load pre-trained TrOCR model
model = VisionEncoderDecoderModel.from_pretrained("/content/drive/MyDrive/OCR-Custom-Dataset/trocr-rec-model")
```

```python
from transformers import Seq2SeqTrainingArguments

training_args = Seq2SeqTrainingArguments(
    output_dir="./trocr-finetuned",          # Output directory
    evaluation_strategy="steps",             # Evaluate during training
    save_strategy="steps",                   # Save model checkpoints
    save_steps=500,                          # Save every 500 steps
    eval_steps=500,                          # Evaluate every 500 steps
    logging_steps=100,                       # Log training metrics
    per_device_train_batch_size=4,           # Batch size for training
    per_device_eval_batch_size=4,            # Batch size for evaluation
    num_train_epochs=3,                      # Number of training epochs
    learning_rate=5e-5,                      # Learning rate
    weight_decay=0.05,                       # Weight decay for regularization
    save_total_limit=2,                      # Limit to 2 checkpoints
    load_best_model_at_end=True,             # Load the best model at the end
    predict_with_generate=True               # Use generate method during eval
)
```

```python
from transformers import Seq2SeqTrainer
# c
# Set the decoder start token ID and pad token ID
model.config.decoder_start_token_id = processor.tokenizer.cls_token_id
model.config.pad_token_id = processor.tokenizer.pad_token_id

# Define Trainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)
```

```python
# Train the model
trainer.train()
```

```python
def custom_recognition_model(img):

  ## Check for gray & high images
  img = np.array(img)
  if len(img.shape) == 2 or len(img.shape) == 4:
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)

  inputs = processor_v1(images=img, return_tensors="pt")
  ## Perform inference
  with torch.no_grad():
      generated_ids = model_v1.generate(**inputs)
  ## Decode the generated text
  generated_text = processor_v1.decode(generated_ids[0], skip_special_tokens=True)

  ## Verify results
  if generated_text is None:
    return img, "None"
  return img, generated_text
```

```python
def detection_img(img_path):
  img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)


  X_dim, Y_dim = img.shape[0]//2048, img.shape[1]//2048
  X_dim, Y_dim = X_dim * 2048, Y_dim * 2048
  img = img[0:X_dim, 0:Y_dim]
  patch_imgs = patchify.patchify(img, (2048, 2048, 3), step=2048)
  patch_imgs = patch_imgs.reshape(-1, 2048, 2048, 3)


  ## run detection on patches
  detection_results = []
  for id, i in enumerate(patch_imgs):
    detect_inference = detection_model(i)
    detection_results.append(detect_inference)

  return detection_results
```

```python
detection_results = detection_img("/content/bw_image_1 (7).jpg")
```

```python
import cv2
import os
output_dir = "/content/output/"
os.makedirs(output_dir, exist_ok=True)
recog_results = []

## Process detection results
for idx, results in enumerate(detection_results):
  for idx1, result in enumerate(results):
    try:
        boxes = result.boxes  # Bounding boxes
        # Check if there are any detections before accessing boxes
        if boxes.shape[0] > 0:  # Check if boxes is not empty
          for box in boxes:
              x1, y1, x2, y2 = box.xyxy[0]  # Bounding box coordinates (x1, y1, x2, y2)
              conf = box.conf[0]  # Confidence score
              cls = int(box.cls[0])  # Class index
              label = detection_model.names[cls]  # Class name

              #################################
              #  RECOGNITION MODEL INFERENCE  #
              #################################
              print("Recognition started")
              crop_img = PILImage.fromarray(result[0].orig_img[int(y1):int(y2), int(x1):int(x2)])
              img, genereted_text = custom_recognition_model(crop_img)
              # ocr_result = ocr.ocr(crop_img, cls=True, det=False, rec=True)
              # genereted_text = ocr_result[0][0][0]
              # print("Generated_1", genereted_text)


              recog_results.append({"img":img, "text":genereted_text})
              print(f"Class: {label}, Confidence: {conf:.2f}, Box: {x1}, {y1}, {x2}, {y2}")

              # Visualize results (optional)
              cv2.rectangle(result[0].orig_img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
              cv2.putText(result[0].orig_img, f"{genereted_text} ", (int(x1), int(y1) - 5),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Save or display the results[0].orig_img
        cv2.imwrite(os.path.join(output_dir,f"{str(idx)}_output.jpg"), result[0].orig_img)
        print(f"Image saved at: {os.path.join(output_dir,f'{str(idx)}_output.jpg')}")
    except Exception as e:
      continue




```
